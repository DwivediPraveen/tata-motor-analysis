```python
tm['PriceDiff'] = tm['Close'].shift(-1) - tm['Close']
```

```python
tm['Return'] = tm['PriceDiff']/tm['Close']
```

```python
tm['Direction'] = [1 if tm['PriceDiff'].loc[ei] > 0 else 0 for ei in tm.index]
```

```python
tm['ma50'] = tm['Close'].rolling(50).mean()

#plot the MA
plt.figure(figsize=(20, 10))
tm['ma50'].loc['2020-01-01':'2020-12-31'].plot(label='MA50')
tm['Close'].loc['2020-01-01':'2020-12-31'].plot(label='Close')
plt.legend()
plt.show()
```

```python
#Strategy
tm = pd.read_csv('C:/Users/pk125/OneDrive/Desktop/code/TATAMOTORS.csv',index_col=0)
tm['MA50'] = tm['Close'].rolling(50).mean()
tm['MA200'] = tm['Close'].rolling(200).mean()
tm = tm.dropna()
tm.head()
```

```python
tm['Shares'] = [1 if tm.loc[ei, 'MA50'] > tm.loc[ei, 'MA200'] else 0 for ei in tm.index]
```

```python
tm['Close1'] = tm['Close'].shift(-1)
tm.iloc[0:616,:]
```

```python
tm['Profit'] = [tm.loc[ei,'Close1']-tm.loc[ei,'Close'] 
                if tm.loc[ei,'Shares']==1
                else 0 for ei in tm.index]
tm['Profit'].plot()
plt.axhline(y=0, color='red')
```

```python
tm['wealth'] = tm['Profit'].cumsum()
tm.tail()
```

```python
tm['wealth'].plot()
plt.title('Total money you win/loss is {}'.format(tm.loc[tm.index[-2], 'wealth']))
```
